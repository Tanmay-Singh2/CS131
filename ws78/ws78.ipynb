```python
from pyspark.sql import SparkSession
import time
import matplotlib.pyplot as plt

runtimes = []
file_sizes = [10, 50, 100, 250, 500]

start = time.time()
df1 = spark.read.option("header", "true").csv("gs://dataproc-staging-us-central1-857864625438-mu4tgbuf/data/data_sample.csv")
df1 = df1.withColumn("value", df1["value"].cast("int"))
df1.groupBy("name").avg("value").show()
end = time.time()
runtimes.append(end - start)
print(end - start, " seconds to run")
```
